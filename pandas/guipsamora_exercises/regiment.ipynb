```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|',index_col='user_id')
```

```python
users.head()
```

```python
users.groupby('occupation').mean()
```

```python
def gender_to_numeric(x):
    if x=='M':
        return 1
    if x=='F':
        return 0
users['gender_count']= users['gender'].apply(gender_to_numeric)  
a = users.groupby('occupation').gender_count.sum()/ users.occupation.value_counts() * 100
a.sort_values(ascending=False)  
    
```

```python
users.groupby('occupation').age.agg(['max','min'])
```

```python
users.groupby(['occupation','gender']).age.mean()
```

```python
gender_count = users.groupby(['occupation','gender']).agg({'gender':'count'})
occupation_count = users.groupby('occupation').agg('count')
ratio = gender_count/occupation_count *100
ratio.loc[ : ,'gender']
```
